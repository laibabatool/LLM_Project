```python
import PyPDF2
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
```

```python
#reading the pdf and extracting text from it, then making a query and through a template running a chain over query and pdf_text
#to get the desired response from llm
```

```python
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text()
    return text

pdf_path ='Laiba Batool data science.pdf'
pdf_text = extract_text_from_pdf(pdf_path)

#reader = PyPDF2.PdfReader('Laiba Batool data science.pdf')

query = "extract the email and contact from the pdf."  # The specific information you want to extract

info_prompt = PromptTemplate.from_template("Extract specific information about {query} from the following text: {pdf_text}")
info_chain = LLMChain(llm=llm, prompt=info_prompt, verbose=True)
resp = info_chain.run(query=query, pdf_text=pdf_text)
#resp = info_chain.run(query=query)
print(resp)
                          
```

```python
import transformers
import torch

model_id = "meta-llama/Meta-Llama-3-70B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)

messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "Who are you?"},
]

prompt = pipeline.tokenizer.apply_chat_template(
        messages, 
        tokenize=False, 
        add_generation_prompt=True
)

terminators = [
    pipeline.tokenizer.eos_token_id,
    pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]

outputs = pipeline(
    prompt,
    max_new_tokens=256,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9,
)
print(outputs[0]["generated_text"][len(prompt):])
```
